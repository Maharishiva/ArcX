Basic marginalisation
$$p(x) = \sum_{y} p(x,y)$$
$$p(y) = \sum_{x} p(x,y)$$

Entropy
$$H(X) = -\sum_{x} p(x)\log_2 p(x)$$

Conditional entropy
$$H(X|Y) = -\sum_{x,y} p(x,y)\log_2 p(x|y)$$
$$= -\sum_{x,y} p(x,y)\log_2\frac{p(x,y)}{p(y)}$$

Mutual information (three equivalent forms)
$$I(X;Y) = \sum_{x,y} p(x,y)\log_2\frac{p(x,y)}{p(x)p(y)}$$
$$= H(X) - H(X|Y)$$
$$= H(Y) - H(Y|X)$$
